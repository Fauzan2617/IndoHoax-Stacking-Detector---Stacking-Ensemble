```python
import pandas as pd
import random

# ============================================================
# 1. LOAD DATA UTAMA
# ============================================================

df = pd.read_csv("../data/raw/datasetUMPOHoax.csv")

print("Jumlah data awal:", len(df))


# ============================================================
# 2. TEMPLATE DATA FAKTA (VARIASI BAGUS)
# ============================================================

subjek = [
    "Pemerintah Indonesia",
    "WHO",
    "Bank Indonesia",
    "Kementerian Kesehatan",
    "Badan Pusat Statistik",
    "Universitas Indonesia",
    "OpenAI",
    "Google",
    "Tesla",
    "Kementerian Pendidikan"
]

aksi = [
    "mengumumkan",
    "melaporkan",
    "menjelaskan",
    "mengembangkan",
    "meluncurkan",
    "meningkatkan",
    "melakukan penelitian",
    "mengatur kebijakan",
    "memantau kondisi",
    "menerapkan program"
]

objek = [
    "kebijakan ekonomi nasional",
    "program kesehatan masyarakat",
    "teknologi kecerdasan buatan",
    "data inflasi terbaru",
    "penyebaran penyakit menular",
    "sistem pendidikan nasional",
    "layanan publik digital",
    "transportasi umum",
    "penelitian ilmiah",
    "kualitas udara di kota besar"
]

keterangan = [
    "untuk meningkatkan kesejahteraan masyarakat",
    "sebagai bagian dari upaya nasional",
    "guna mendukung pertumbuhan ekonomi",
    "untuk mengatasi permasalahan global",
    "sebagai langkah preventif",
    "dalam rangka inovasi teknologi",
    "untuk meningkatkan efisiensi",
    "demi kepentingan publik",
    "sebagai solusi jangka panjang",
    "untuk menghadapi tantangan masa depan"
]


# ============================================================
# 3. GENERATE DATA FAKTA
# ============================================================

def generate_data(n=1500):
    data = []

    for _ in range(n):
        kalimat = f"{random.choice(subjek)} {random.choice(aksi)} {random.choice(objek)} {random.choice(keterangan)}."
        data.append((kalimat, "fakta"))  # pakai label string biar konsisten awal

    return pd.DataFrame(data, columns=["tweet", "label"])


df_generated = generate_data(1500)

print("Data fakta tambahan:", len(df_generated))


# ============================================================
# 4. TAMBAH VARIASI (GAYA REAL / TWITTER / INFORMAL)
# ============================================================

data_variatif = [
    ("Pemerintah lagi fokus ningkatin ekonomi nasional biar stabil", "fakta"),
    ("WHO lagi mantau penyebaran penyakit global", "fakta"),
    ("AI sekarang dipakai di banyak bidang kerja", "fakta"),
    ("Bank Indonesia lagi jaga inflasi tetap terkendali", "fakta"),
    ("Transportasi umum bantu mobilitas masyarakat sehari-hari", "fakta"),
    ("Mahasiswa belajar berbagai ilmu di kampus", "fakta"),
    ("Teknologi berkembang pesat dalam beberapa tahun terakhir", "fakta"),
    ("Informasi bisa diakses dengan mudah lewat internet", "fakta"),
]

df_variatif = pd.DataFrame(data_variatif, columns=["tweet", "label"])


# ============================================================
# 5. GABUNGKAN SEMUA DATA
# ============================================================

df_final = pd.concat([df, df_generated, df_variatif], ignore_index=True)

print("Jumlah setelah digabung:", len(df_final))


# ============================================================
# 6. SHUFFLE (PENTING BIAR ACAK)
# ============================================================

df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)


# ============================================================
# 7. SIMPAN DATA
# ============================================================

df_final.to_csv("../data/raw/dataset_gabungan.csv", index=False)

print("✅ Dataset gabungan berhasil dibuat!")
```
