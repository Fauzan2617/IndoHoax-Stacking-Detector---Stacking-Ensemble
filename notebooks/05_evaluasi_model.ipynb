```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from src.utils import load_objek

# ============================================================
# 1. LOAD DATA & MODEL
# ============================================================

data = load_objek('../data/processed/data_vektor.pkl')
model = load_objek('../models/stacking_final_tuning.pkl')

X_test = data['X_test_clean']
y_test = data['y_test']

# ============================================================
# 2. PREDIKSI
# ============================================================

y_pred = model.predict(X_test)

# ============================================================
# 3. CONFUSION MATRIX
# ============================================================

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

# ============================================================
# 4. VISUALISASI (LEBIH BAGUS BUAT LAPORAN)
# ============================================================

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Fakta (0)', 'Hoaks (1)'],
    yticklabels=['Fakta (0)', 'Hoaks (1)']
)

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Model Stacking')

plt.tight_layout()

# Simpan gambar
plt.savefig('../data/confusion_matrix.png')

plt.show()
```
