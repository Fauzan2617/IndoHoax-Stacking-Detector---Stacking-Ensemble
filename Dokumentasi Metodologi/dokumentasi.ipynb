```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
import re

# =====================================================
# CONTOH UNTUK GAMBAR LAPORAN
# =====================================================

teks = "Waspada vaksin covid mengandung microchip 5g tahun baca di"

print("INPUT TEKS:")
print(teks)


# =========================
# 1. CASE FOLDING + CLEANING
# =========================

teks = teks.lower()

teks = re.sub(r'[^a-zA-Z\s]', '', teks)

print("\nSETELAH CLEANING:")
print(teks)


# =========================
# 2. TOKENIZING
# =========================

tokens = teks.split()

print("\nHASIL TOKENIZING:")
print(tokens)


# =========================
# 3. STOPWORD REMOVAL
# =========================

factory = StopWordRemoverFactory()

stopwords = set(factory.get_stop_words())

# tambah custom noise kalau perlu
stopwords.update(['g', 'tahun'])

filtered_tokens = [
    word for word in tokens
    if word not in stopwords
]

print("\nHASIL STOPWORD REMOVAL:")
print(filtered_tokens)
```
